# Student Marks Prediction using Linear Regression


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Create dataset
data = {
    "Student": ["Aman", "Rahul", "Sara", "Fatima", "John"],
    "Study_Hours": [2, 5, 3, 8, 6],
    "Marks": [50, 75, 60, 90, 85]
}

df = pd.DataFrame(data)

df
```

```python
df.head()
```

```python
df.describe()
```

```python
df["Marks"].mean()
```

```python
#Data visualization
plt.bar(df["Student"], df["Marks"])
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Student Marks Analysis")
plt.show()
```

```python
plt.scatter(df["Study_Hours"], df["Marks"])
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[["Study_Hours"]]   # input (must be 2D)
y = df["Marks"]           # output

print(X)
print(y)
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
prediction = model.predict([[7]])
print("Predicted marks for 7 study hours:", prediction)
```

```python
plt.scatter(df["Study_Hours"], df["Marks"])

predicted_line = model.predict(X)
plt.plot(df["Study_Hours"], predicted_line, color="red")

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Linear Regression Model")
plt.show()
```

```python
print("Model Score (R^2):", model.score(X, y))

```

```python
from sklearn.model_selection import train_test_split

X = df[["Study_Hours"]]
y = df["Marks"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=42
)

print("Training size:", len(X_train))
print("Testing size:", len(X_test))
```

```python
#Train model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#Test model
predictions = model.predict(X_test)

print("Actual marks:", y_test.values)
print("Predicted marks:", predictions)

```

```python
#r^2
print("Test R^2 Score:", model.score(X_test, y_test))
```
